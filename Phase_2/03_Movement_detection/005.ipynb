```python
import cv2
import numpy as np
import os
import math
import json
```

```python
# Contour parameters
min_contour_width = 25  
min_contour_height = 25
max_contour_width = 150 
max_contour_height = 150
# Distance from the line
offset = 20
# Lane parameters
lane_1_height_1=210
lane_1_height_2=230
lane_2_height_1=240
lane_2_height_2=260
lane_1_width_1=40
lane_1_width_2=75
lane_2_width_1=90
lane_2_width_2=130
lane_3_height_1=260
lane_3_height_2=270
lane_4_height_1=240
lane_4_height_2=250
lane_3_width_1=230
lane_3_width_2=195
lane_4_width_1=290
lane_4_width_2=250
matches = []
# Vehicle counters
vehicles_3 = 0
vehicles_2 = 0
# Other helping metrics
frame_number_threshold=15
last_1=""
last_2=""
last_3=""
last_4=""
new_x_2=0
new_y_2=0
new_x_3=0
new_y_3=0
```

```python
# Line to point distance function
def point_to_line_distance(point, line_start_width, line_start_height, line_end_width, line_end_height):

    A = line_end_height - line_start_height
    B = line_start_width - line_end_width
    C = line_end_width * line_start_height - line_start_width * line_end_height

    distance = abs(A * point[0] + B * point[1] + C) / np.sqrt(A**2 + B**2)
    return distance
```

```python
# Point to point distance function
def distance_between_points(x1, y1, x2, y2):

    distance = math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)
    return distance
```

```python
# Getting centroid
def get_centrolid(x, y, w, h):
   x1 = int(w / 2)
   y1 = int(h / 2)
 
   cx = x + x1
   cy = y + y1
   return cx, cy
```

```python
results={}

# Iterating through videos

for video in files:

    # Defining counters
    
    vehicles_1 = 0
    vehicles_2 = 0
    vehicles_3 = 0
    vehicles_4 = 0

    motorbikes_1=0
    motorbikes_2=0
    motorbikes_3=0
    motorbikes_4=0

    cars_1=0
    cars_2=0
    cars_3=0
    cars_4=0

    transporters_1=0
    transporters_2=0
    transporters_3=0
    transporters_4=0

    trucks_1=0
    trucks_2=0
    trucks_3=0
    trucks_4=0

    motorbikes_1_list=[]
    motorbikes_2_list=[]
    motorbikes_3_list=[]
    motorbikes_4_list=[]

    cars_1_list=[]
    cars_2_list=[]
    cars_3_list=[]
    cars_4_list=[]

    transporters_1_list=[]
    transporters_2_list=[]
    transporters_3_list=[]
    transporters_4_list=[]

    trucks_1_list=[]
    trucks_2_list=[]
    trucks_3_list=[]
    trucks_4_list=[]

    # Loading videos
    
    cap = cv2.VideoCapture(video)
    
    cap.set(3, 620)
    cap.set(4, 480)
     
    if cap.isOpened():
       ret, frame1 = cap.read()
    else:
       ret = False
    ret, frame1 = cap.read()
    ret, frame2 = cap.read()
    
    frame_nr=0
    frame_1_counter=0
    frame_2_counter=0
    frame_3_counter=0
    frame_4_counter=0
    
    while ret:
    
       frame_1_counter+=1
       frame_2_counter+=1
       frame_3_counter+=1
       frame_4_counter+=1

       # Frame difference calculations and preprocessing
    
       d = cv2.absdiff(frame1, frame2)
       grey = cv2.cvtColor(d, cv2.COLOR_BGR2GRAY)
     
       blur = cv2.GaussianBlur(grey, (5, 5), 0)
     
       ret, th = cv2.threshold(blur, 20, 255, cv2.THRESH_BINARY)
       dilated = cv2.dilate(th, np.ones((3, 3)))
       kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (2, 2))
     
      
       closing = cv2.morphologyEx(dilated, cv2.MORPH_CLOSE, kernel)
       contours, h = cv2.findContours(
           closing, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
       for(i, c) in enumerate(contours):
           (x, y, w, h) = cv2.boundingRect(c)
           contour_valid = (w >= min_contour_width) and (h >= min_contour_height) and (h <= max_contour_height) and (w <= max_contour_width)
     
           if not contour_valid:
               continue

           # Drawing rectangles over centroids
           
           cv2.rectangle(frame1, (x-10, y-10), (x+w+10, y+h+10), (255, 0, 0), 2)

           # Drawing lines
     
           cv2.line(frame1, (lane_1_width_1, lane_1_height_1), (lane_1_width_2, lane_1_height_2), (0, 0, 255), 5)
           cv2.line(frame1, (lane_2_width_1, lane_2_height_1), (lane_2_width_2, lane_2_height_2), (0, 0, 255), 5)
           cv2.line(frame1, (lane_3_width_1, lane_3_height_1), (lane_3_width_2, lane_3_height_2), (0, 255, 0), 5)
           cv2.line(frame1, (lane_4_width_1, lane_4_height_1), (lane_4_width_2, lane_4_height_2), (0, 255, 0), 5)
           centrolid = get_centrolid(x, y, w, h)
           matches.append(centrolid)
           cv2.circle(frame1, centrolid, 5, (0, 255, 0), -1)
           cx, cy = get_centrolid(x, y, w, h)

           # Checking if each point in within the given distance from the line, on right width and classifying vehicle type
           
           for (x, y) in matches:

               if x < lane_1_width_2+4 and x > lane_1_width_1-3 and point_to_line_distance((x, y), lane_1_width_1, lane_1_height_1, lane_1_width_2, lane_1_height_2)<offset:
                   if frame_1_counter<=frame_number_threshold and frame_1_counter>12:
                       if last_1=="motorbike":
                           motorbikes_1-=1
                           try:
                               motorbikes_1_list.pop()
                           except:
                               pass
                               
                       if last_1=="car":
                           cars_1-=1
                           try:
                               cars_1_list.pop()
                           except:
                               pass
                               
                       if last_1=="transporter":
                           transporters_1-=1
                           try:
                               transporters_1_list.pop()
                           except:
                               pass
                               
                       if last_1=="truck":
                           trucks_1-=1
                           try:
                               trucks_1_list.pop()
                           except:
                               pass
                       trucks_1+=1
                       trucks_1_list.append(frame_nr)
                           
                   elif frame_1_counter<=12 and frame_1_counter>10:
                       if last_1=="motorbike":
                           motorbikes_1-=1
                           try:
                               motorbikes_1_list.pop()
                           except:
                               pass
                           
                       if last_1=="car":
                           cars_1-=1
                           try:
                               cars_1_list.pop()
                           except:
                               pass
                               
                       if last_1=="transporter":
                           transporters_1-=1
                           try:
                               transporters_1_list.pop()
                           except:
                               pass
                           
                       if last_1=="truck":
                           trucks_1-=1
                           try:
                               trucks_1_list.pop()
                           except:
                               pass
                       transporters_1+=1
                       transporters_1_list.append(frame_nr)
                           
                   elif frame_1_counter>frame_number_threshold:
                       if w<=30 and h<=30:
                           motorbikes_1+=1
                           motorbikes_1_list.append(frame_nr)
                           last_1="motorbike"
                       elif w>30 and h>30 and w<75 and h<75:
                           cars_1+=1
                           cars_1_list.append(frame_nr)
                           last_1="car"
                       elif w<75 and h<75 and w>=100 and h>=100:
                           transporters_1+=1
                           transporters_1_list.append(frame_nr)
                           last_1="transporter"
                       elif w>=80 and h>=80:
                           trucks_1+=1
                           trucks_1_list.append(frame_nr)
                           last_1="truck"
                       else:
                           cars_1+=1
                           cars_1_list.append(frame_nr)
                           last_1="car"
                       vehicles_1+=1
                   frame_1_counter=0
                   matches.remove((x, y))
               
               if x < lane_2_width_2+3 and x > lane_2_width_1-3 and point_to_line_distance((x, y), lane_2_width_1, lane_2_height_1, lane_2_width_2, lane_2_height_2)<offset:
                   if frame_2_counter<=frame_number_threshold and frame_2_counter>12:
                       if last_2=="motorbike":
                           motorbikes_2-=1
                           try:
                               motorbikes_2_list.pop()
                           except:
                               pass
                           
                       if last_2=="car":
                           cars_2-=1
                           try:
                               cars_2_list.pop()
                           except:
                               pass
                           
                       if last_2=="transporter":
                           transporters_2-=1
                           try:
                               transporters_2_list.pop()
                           except:
                               pass
                           
                       if last_2=="truck":
                           trucks_2-=1
                           try:
                               trucks_2_list.pop()
                           except:
                               pass
                       trucks_2+=1
                       trucks_2_list.append(frame_nr)
                           
                   elif frame_2_counter<=12 and frame_2_counter>10:
                       if last_2=="motorbike":
                           motorbikes_2-=1
                           try:
                               motorbikes_2_list.pop()
                           except:
                               pass
                           
                       if last_2=="car":
                           cars_2-=1
                           try:
                               cars_2_list.pop()
                           except:
                               pass
                           
                       if last_2=="transporter":
                           transporters_2-=1
                           try:
                               transporters_2_list.pop()
                           except:
                               pass
                           
                       if last_2=="truck":
                           trucks_2-=1
                           try:
                               trucks_2_list.pop()
                           except:
                               pass
                       transporters_2+=1
                       transporters_2_list.append(frame_nr)
                   elif frame_2_counter>frame_number_threshold:
                       if w<=35 and h<=35:
                           motorbikes_2+=1
                           motorbikes_2_list.append(frame_nr)
                           last_2="motorbike"
                       elif w>35 and h>35 and w<75 and h<75:
                           cars_2+=1
                           cars_2_list.append(frame_nr)
                           last_2="car"
                       elif w<100 and h<100 and w>=75 and h>=75:
                           transporters_2+=1
                           transporters_2_list.append(frame_nr)
                           last_2="transporter"
                       elif w>=100 and h>=100:
                           trucks_2+=1
                           trucks_2_list.append(frame_nr)
                           last_2="truck"
                       else:
                           cars_2+=1
                           cars_2_list.append(frame_nr)
                           last_2="car"
                       vehicles_2+=1
                   frame_2_counter=0
                   matches.remove((x, y))
                   
               if x > lane_4_width_2-3 and x < lane_4_width_1+3 and point_to_line_distance((x, y), lane_4_width_1, lane_4_height_1, lane_4_width_2, lane_4_height_2)<offset:
                   if frame_4_counter<=frame_number_threshold and frame_4_counter>12:
                       if last_4=="motorbike":
                           motorbikes_4-=1
                           try:
                               motorbikes_4_list.pop()
                           except:
                               pass
                           
                       if last_4=="car":
                           cars_4-=1
                           try:
                               cars_4_list.pop()
                           except:
                               pass
                           
                       if last_4=="transporter":
                           transporters_4-=1
                           try:
                               transporters_4_list.pop()
                           except:
                               pass
                           
                       if last_4=="truck":
                           trucks_4-=1
                           try:
                               trucks_4_list.pop()
                           except:
                               pass
                       trucks_4+=1
                       trucks_4_list.append(frame_nr)
                   elif frame_4_counter<=12 and frame_4_counter>10:
                       if last_4=="motorbike":
                           motorbikes_4-=1
                           try:
                               motorbikes_4_list.pop()
                           except:
                               pass
                           
                       if last_4=="car":
                           cars_4-=1
                           try:
                               cars_4_list.pop()
                           except:
                               pass
                           
                       if last_4=="transporter":
                           transporters_4-=1
                           try:
                               transporters_4_list.pop()
                           except:
                               pass
                           
                       if last_4=="truck":
                           trucks_4-=1
                           try:
                               trucks_4_list.pop()
                           except:
                               pass
                       transporters_4+=1
                       transporters_4_list.append(frame_nr)
                   elif frame_4_counter>frame_number_threshold:
                       if w<=35 and h<=35:
                           motorbikes_4+=1
                           motorbikes_4_list.append(frame_nr)
                           last_4="motorbike"
                       elif w>35 and h>35 and w<70 and h<70:
                           cars_4+=1
                           cars_4_list.append(frame_nr)
                           last_4="car"
                       elif w<100 and h<100 and w>=70 and h>=70:
                           transporters_4+=1
                           transporters_4_list.append(frame_nr)
                           last_4="transporter"
                       elif w>=100 and h>=100:
                           trucks_4+=1
                           trucks_4_list.append(frame_nr)
                           last_4="truck"
                       else:
                           cars_4+=1
                           cars_4_list.append(frame_nr)
                           last_4="car"
                       vehicles_4+=1
                   frame_4_counter=0
                   matches.remove((x, y))
                   
               if x < lane_3_width_1+3 and x > lane_3_width_2-3 and point_to_line_distance((x, y), lane_3_width_1, lane_3_height_1, lane_3_width_2, lane_3_height_2)<offset:
                   if frame_3_counter<=frame_number_threshold and frame_3_counter>12:
                       if last_3=="motorbike":
                           motorbikes_3-=1
                           try:
                               motorbikes_3_list.pop()
                           except:
                               pass
                           
                       if last_3=="car":
                           cars_3-=1
                           try:
                               cars_3_list.pop()
                           except:
                               pass
                           
                       if last_3=="transporter":
                           transporters_3-=1
                           try:
                               transporters_3_list.pop()
                           except:
                               pass
                           
                       if last_3=="truck":
                           trucks_3-=1
                           try:
                               trucks_3_list.pop()
                           except:
                               pass
                       trucks_3+=1
                       trucks_3_list.append(frame_nr)
                       
                   elif frame_3_counter<=12 and frame_3_counter>10:
        
                       if last_3=="motorbike":
                           motorbikes_3-=1
                           try:
                               motorbikes_3_list.pop()
                           except:
                               pass
                           
                       if last_3=="car":
                           cars_3-=1
                           try:
                               cars_3_list.pop()
                           except:
                               pass
                           
                       if last_3=="transporter":
                           transporters_3-=1
                           try:
                               transporters_3_list.pop()
                           except:
                               pass
                           
                       if last_3=="truck":
                           trucks_3-=1
                           try:
                               trucks_3_list.pop()
                           except:
                               pass
                       transporters_3+=1
                       transporters_3_list.append(frame_nr)
                       
                   elif frame_3_counter>frame_number_threshold:
                       if w<=40 and h<=40:
                           motorbikes_3+=1
                           motorbikes_3_list.append(frame_nr)
                           last_3="motorbike"
                       elif w>40 and h>40 and w<70 and h<70:
                           cars_3+=1
                           cars_3_list.append(frame_nr)
                           last_3="car"
                       elif w<100 and h<100 and w>=70 and h>=70:
                           transporters_3+=1
                           transporters_3_list.append(frame_nr)
                           last_3="transporter"
                       elif w>=100 and h>=100:
                           trucks_3+=1
                           trucks_3_list.append(frame_nr)
                           last_3="truck"
                       else:
                           cars_3+=1
                           cars_3_list.append(frame_nr)
                           last_3="car"
                       vehicles_3+=1
                   frame_3_counter=0
                   matches.remove((x, y))

       # Text with current counts on the video
     
       cv2.putText(frame1, f"Total Vehicle Detected Lane 1: {vehicles_1}", (0, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.6,
                   (0, 170, 0), 2)
       cv2.putText(frame1, f"Motorbikes: {motorbikes_1}, cars: {cars_1}, transporters: {transporters_1}, trucks: {trucks_1}", (0, 50), cv2.FONT_HERSHEY_SIMPLEX, 0.4,
                   (170, 0, 0), 1)
       cv2.putText(frame1, f"Total Vehicle Detected Lane 2: {vehicles_2}", (0, 70), cv2.FONT_HERSHEY_SIMPLEX, 0.6,
                   (0, 170, 0), 2)
       cv2.putText(frame1, f"Motorbikes: {motorbikes_2}, cars: {cars_2}, transporters: {transporters_2}, trucks: {trucks_2}", (0, 90), cv2.FONT_HERSHEY_SIMPLEX, 0.4,
                   (170, 0, 0), 1)
       cv2.putText(frame1, f"Total Vehicle Detected Lane 3: {vehicles_3}", (0, 110), cv2.FONT_HERSHEY_SIMPLEX, 0.6,
                   (0, 170, 0), 2)
       cv2.putText(frame1, f"Motorbikes: {motorbikes_3}, cars: {cars_3}, transporters: {transporters_3}, trucks: {trucks_3}", (0, 130), cv2.FONT_HERSHEY_SIMPLEX, 0.4,
                   (170, 0, 0), 1)
       cv2.putText(frame1, f"Total Vehicle Detected Lane 4: {vehicles_4}", (0, 150), cv2.FONT_HERSHEY_SIMPLEX, 0.6,
                   (0, 170, 0), 2)
       cv2.putText(frame1, f"Motorbikes: {motorbikes_4}, cars: {cars_4}, transporters: {transporters_4}, trucks: {trucks_4}", (0, 170), cv2.FONT_HERSHEY_SIMPLEX, 0.4,
                   (170, 0, 0), 1)
       cv2.imshow("Vehicle Detection", frame1)
        
       if cv2.waitKey(1) == 27:
           break
       # Wait for a specified amount of time
       if cv2.waitKey(20) & 0xFF == ord('q'):
            break
       frame1 = frame2
       ret, frame2 = cap.read()
       frame_nr+=1
     
    cv2.destroyAllWindows()
    cap.release()

    # Saving results into dictionary

    results[video]={"motorbikes_lane_1":motorbikes_1, "motorbikes_lane_1_frames":motorbikes_1_list, "cars_lane_1":cars_1, "cars_lane_1_frames":cars_1_list, "transporters_lane_1":transporters_1, "transporters_lane_1_frames":transporters_1_list, "trucks_lane_1":trucks_1, "trucks_lane_1_frames":trucks_1_list,
    "motorbikes_lane_2":motorbikes_2, "motorbikes_lane_2_frames":motorbikes_2_list, "cars_lane_2":cars_2, "cars_lane_2_frames":cars_2_list, "transporters_lane_2":transporters_2, "transporters_lane_2_frames":transporters_2_list, "trucks_lane_2":trucks_2, "trucks_lane_2_frames":trucks_2_list,
    "motorbikes_lane_3":motorbikes_3, "motorbikes_lane_3_frames":motorbikes_3_list, "cars_lane_3":cars_3, "cars_lane_3_frames":cars_3_list, "transporters_lane_3":transporters_3, "transporters_lane_3_frames":transporters_3_list, "trucks_lane_3":trucks_3, "trucks_lane_3_frames":trucks_3_list,
    "motorbikes_lane_4":motorbikes_4, "motorbikes_lane_4_frames":motorbikes_4_list, "cars_lane_4":cars_4, "cars_lane_4_frames":cars_4_list, "transporters_lane_4":transporters_4, "transporters_lane_4_frames":transporters_4_list, "trucks_lane_4":trucks_4, "trucks_lane_4_frames":trucks_4_list}
```
